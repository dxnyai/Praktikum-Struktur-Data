**Representasi Graf dengan Adjacency List**

```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Menampilkan graf
def print_graph(graf):
  for node in graf:
    print(f"{node} -> {graf[node]}")

print_graph(graph)
```

**Traversal Graf menggunakan BFS**

```python
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Contoh pemanggilan
bfs(graph, 'A')
```

**Traversal Graf menggunakan DFS**

```python
def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

# Contoh pemanggilan
dfs(graph, 'A')
```

**Latihan Personal**

```python
# Lengkapi fungsi untuk menghitung jumlah tetangga dari tiap simpul dalam graf

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

def count_neighbors(graph):
    result = {}

    for node in graph:
        result[node] = len(graph[node])

    return result

print(count_neighbors(graph))
```

```python
# 1. Buatlah graf tak berarah baru dengan minimal 5 simpul dan 7 sisi, lalu tampilkan dengan fungsi print_graph.
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C', 'E'],
    'C': ['A', 'B', 'D', 'F'],
    'D': ['A', 'C', 'E'],
    'E': ['B', 'D', 'F'],
    'F': ['C', 'E']
}

def print_graph(graph):
    for node in graph:
        print(f"{node} -> {graph[node]}")

print("Graph:")
print_graph(graph)


# 2. Implementasikan BFS dan DFS dari simpul awal pilihan Anda.

# Implementasi BFS
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        vertex = queue.popleft()

        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)

print("\n\nHasil BFS:")
bfs(graph, 'A')


# Implementasi DFS
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        print(start, end=" ")
        visited.add(start)

        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

print("\n\nHasil DFS:")
dfs(graph, 'A')


# 3. Modifikasi fungsi BFS agar mengembalikan list urutan kunjungan, bukan hanya mencetak.
def bfs_list(graph, start):
    visited = set()
    queue = deque([start])
    result = []

    while queue:
        vertex = queue.popleft()

        if vertex not in visited:
            visited.add(vertex)
            result.append(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)

    return result

print("\n\nList Urutan BFS:")
print(bfs_list(graph, 'A'))


# 4. Buat fungsi find_path(graph, start, end) yang mengembalikan satu jalur dari start ke end jika ada, menggunakan DFS.
def find_path(graph, start, end, path=None):
    if path is None:
        path = []

    path = path + [start]

    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path(graph, neighbor, end, path)
            if new_path:
                return new_path

    return None

print("\nPath dari A ke F:")
print(find_path(graph, 'A', 'F'))


# 5. Tambahkan fungsi is_connected(graph) untuk mengecek apakah semua simpul dalam graf saling terhubung (gunakan kombinasi DFS + logika)
def is_connected(graph):
    start = next(iter(graph))
    visited = set()

    def dfs_check(node):
        visited.add(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_check(neighbor)

    dfs_check(start)

    return len(visited) == len(graph)

print("\nApakah graph terhubung?")
print(is_connected(graph))
```
