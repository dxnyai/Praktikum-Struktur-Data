**Binary Tree Manual**

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

print("Hasil inorder traversal dari binary tree: ")
inorder(root)
```

**Binary Search Tree**

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree:
  def __init__(self):
    self.root = None
  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
  root = bst.insert(root, value)
```

**Traversal pada BST**

```python
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

print("Inorder Traversal: ")
inorder(root)
print("\nPreorder Traversal: ")
preorder(root)
print("\nPostorder Traversal: ")
postorder(root)
```

Pencarian Nilai dalam BST

```python
def search(node, key):
  if node is None or node.data == key:
    return node
  if key < node.data:
    return search(node.left, key)
  return search(node.right, key)

# Uji pencarian
key = 60
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
```

```python
# Struktur Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Membuat binary tree manual
root = Node(20)      # Jumlah huruf nama lengkap

# Anak kiri dan kanan root
root.left = Node(7)    # Jumlah huruf vokal
root.right = Node(62)  # Dua digit terakhir NPM

# Level berikutnya
root.left.left = Node(16)   # Tanggal lahir (DD)
root.right.right = Node(1)  # Bulan lahir (MM)

# Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

# Menampilkan hasil traversal
print("Inorder Traversal:")
inorder(root)
```

**Tugas 2: Program Binary Search Tree (BST)**

```python
# Program Binary Search Tree (BST)

# Membuat class node
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Membuat class BST
class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Fungsi insert data ke BST
    def insert(self, root, data):

        # Jika node kosong
        if root is None:
            return Node(data)

        # Jika data lebih kecil, masuk kiri
        if data < root.data:
            root.left = self.insert(root.left, data)

        # Jika data lebih besar, masuk kanan
        else:
            root.right = self.insert(root.right, data)

        return root

# Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

# Fungsi preorder traversal
def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

# Fungsi postorder traversal
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

# Fungsi pencarian data
def search(node, key):

    # Jika node kosong atau data ditemukan
    if node is None or node.data == key:
        return node

    # Cari ke kiri jika key lebih kecil
    if key < node.data:
        return search(node.left, key)

    # Cari ke kanan jika key lebih besar
    return search(node.right, key)

# Data BST
# 062 = tiga digit terakhir NPM
# 16 = tanggal lahir
# 01 = bulan lahir
# 68 dan 97 = ASCII huruf D dan a
# 6 dan 7 = angka bebas

data_list = [62, 16, 1, 68, 97, 6, 7]

# Membuat BST
bst = BinarySearchTree()
root = None

# Memasukkan data ke BST
for value in data_list:
    root = bst.insert(root, value)

# Menampilkan traversal
print("Inorder Traversal:")
inorder(root)

print("\nPreorder Traversal:")
preorder(root)

print("\nPostorder Traversal:")
postorder(root)

# Pencarian data yang ada
key = 62
result = search(root, key)

if result:
    print(f"\n{key} ditemukan dalam tree.")
else:
    print(f"\n{key} tidak ditemukan.")

# Pencarian data yang tidak ada
key = 50
result = search(root, key)

if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")
```
